# ApexPlanet Data Analytics - Task 4
**Data Storytelling & Statistical Validation**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Ensure charts directory exists
os.makedirs('charts', exist_ok=True)

df = pd.read_csv('Cleaned_Sales_Dataset.csv')
```

## PART 1: Data Understanding

```python
print(f"Number of Rows: {len(df)}")
print(f"Number of Columns: {len(df.columns)}")
print("\nData Types:\n", df.dtypes)
```

## PART 2: Statistical Analysis

```python
stats_cols = ['Total_Sales', 'Quantity', 'Unit_Price', 'Age']
desc_stats = df[stats_cols].describe()
# Adding Variance and Mode
desc_stats.loc['var'] = df[stats_cols].var()
desc_stats.loc['mode'] = df[stats_cols].mode().iloc[0]
desc_stats
```

## PART 3: Category Analysis

```python
cat_rev = df.groupby('Category')['Total_Sales'].sum().reset_index()
print("Top Performing Category:\n", cat_rev.loc[cat_rev['Total_Sales'].idxmax()])
print("\nLowest Performing Category:\n", cat_rev.loc[cat_rev['Total_Sales'].idxmin()])

plt.figure(figsize=(10,6))
plt.bar(cat_rev['Category'], cat_rev['Total_Sales'], color='skyblue')
plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('charts/revenue_by_category.png')
plt.show()
```

## PART 4: Customer Segmentation

```python
def segment_age(age):
    if 18 <= age <= 30: return 'Young Adults'
    elif 31 <= age <= 45: return 'Adults'
    elif 46 <= age <= 60: return 'Middle Age'
    else: return 'Seniors'

df['Segment'] = df['Age'].apply(segment_age)

seg_rev = df.groupby('Segment')['Total_Sales'].sum().reset_index()
total_rev = seg_rev['Total_Sales'].sum()
seg_rev['Contribution_Pct'] = (seg_rev['Total_Sales'] / total_rev) * 100
print(seg_rev)

plt.figure(figsize=(8,6))
plt.pie(seg_rev['Total_Sales'], labels=seg_rev['Segment'], autopct='%1.1f%%', startangle=140)
plt.title('Revenue by Customer Segment')
plt.savefig('charts/revenue_by_segment.png')
plt.show()
```

## PART 5: City Analysis

```python
city_rev = df.groupby('City')['Total_Sales'].sum().reset_index().sort_values(by='Total_Sales', ascending=False)
print("Top Performing City:\n", city_rev.iloc[0])
print("\nLowest Performing City:\n", city_rev.iloc[-1])

plt.figure(figsize=(12,6))
plt.bar(city_rev['City'], city_rev['Total_Sales'], color='lightgreen')
plt.title('Revenue by City')
plt.xlabel('City')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('charts/revenue_by_city.png')
plt.show()
```

## PART 6: Time Series Analysis

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Month'] = df['Order_Date'].dt.to_period('M')
monthly_rev = df.groupby('Month')['Total_Sales'].sum().reset_index()
monthly_rev['Month'] = monthly_rev['Month'].astype(str)

plt.figure(figsize=(12,6))
plt.plot(monthly_rev['Month'], monthly_rev['Total_Sales'], marker='o', linestyle='-', color='coral')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('charts/monthly_revenue_trend.png')
plt.show()
```
